# **Scatter Plot**

sepal_length vs sepal_width

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset('iris')

# Create scatter plot with species color-coding
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',  # Color by species
    palette='viridis',  # Color scheme
    s=100  # Marker size
)

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width (Colored by Species)')

# Show the plot
plt.show()
```

# **HIstogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset('iris')

# Set up the figure and subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle('Histograms of Iris Dataset Feature Pairs', fontsize=16)

# Plot histograms for each pair
# 1. sepal_length vs sepal_width
sns.histplot(data=iris, x='sepal_length', y='sepal_width', ax=axes[0, 0], bins=15, cmap='viridis')
axes[0, 0].set_title('Sepal Length vs Sepal Width')

# 2. sepal_length vs petal_length
sns.histplot(data=iris, x='sepal_length', y='petal_length', ax=axes[0, 1], bins=15, cmap='viridis')
axes[0, 1].set_title('Sepal Length vs Petal Length')

# 3. sepal_length vs petal_width
sns.histplot(data=iris, x='sepal_length', y='petal_width', ax=axes[1, 0], bins=15, cmap='viridis')
axes[1, 0].set_title('Sepal Length vs Petal Width')

# 4. sepal_width vs petal_length
sns.histplot(data=iris, x='sepal_width', y='petal_length', ax=axes[1, 1], bins=15, cmap='viridis')
axes[1, 1].set_title('Sepal Width vs Petal Length')

# 5. sepal_width vs petal_width
sns.histplot(data=iris, x='sepal_width', y='petal_width', ax=axes[2, 0], bins=15, cmap='viridis')
axes[2, 0].set_title('Sepal Width vs Petal Width')

# 6. petal_length vs petal_width
sns.histplot(data=iris, x='petal_length', y='petal_width', ax=axes[2, 1], bins=15, cmap='viridis')
axes[2, 1].set_title('Petal Length vs Petal Width')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

# **Box plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset('iris')

# Set up the figure
plt.figure(figsize=(12, 6))

# Box plot for all numerical features
plt.subplot(1, 2, 1)
sns.boxplot(data=iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
plt.title('Outliers in All Features')
plt.xticks(rotation=45)

# Box plot grouped by species
plt.subplot(1, 2, 2)
sns.boxplot(data=iris, x='species', y='petal_length', palette='Set2')
plt.title('Petal Length Outliers by Species')

plt.tight_layout()
plt.show()
```
